```python
categorical_vars = []
continuous_vars = []
for col in train:
    if col in ("tenure","monthly_charges","total_charges"):
        continuous_vars.append(col)
    else:
        categorical_vars.append(col)
```

```python
categorical_vars
```

```python
continuous_vars
```

```python
def plot_categorical_and_continuous_vars(train,categorical,continuous):
    cols = len(categorical)
    for cont_col in continuous:
        _,ax = plt.subplots(nrows = 1, ncols = cols,sharey = True)
        for i,cat in enumerate(categorical):
            sns.scatterplot(train,x = cont_col,y = cat)
            ax[i].set_xlabel("")
            ax[i].set_ylabel(continuous)
            ax[i].set_title(categorical)
            plt.show()
```

```python
plot_categorical_and_continuous_vars(train, categorical_vars,continuous_vars)
```

```python
def telco_churn_heatmap(df):
    '''
    Makes a heatmap of the data
    '''
    plt.figure(figsize=(8,12))
    churn_heatmap = sns.heatmap(df.corr()[["churn"]].sort_values(by = "churn",ascending = False),vmin = -.5,vmax = .5, annot = True, cmap = "plasma")
    churn_heatmap.set_title("Features Correlation with Churn")
    return churn_heatmap
```
